```python
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib as plt
import statsmodels.api as sm


reading_xlxs = pd.read_excel(open('excel/harry_potter_analysis.xlsx', 'rb'), sheet_name='harry_potter_characters') 

#print(reading_xlxs.head(5)) # If you want show more items, change the paramether.
reading_xlxs_hp_male = reading_xlxs[reading_xlxs["Gender"] == "male"]
reading_xlxs_hp_female = reading_xlxs[reading_xlxs["Gender"] == "female"]

# MALES
male_characters_hair_bald = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "bald"])
male_characters_hair_black = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "black"])
male_characters_hair_blond = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "blond"])
male_characters_hair_brown = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "brown"])
male_characters_hair_dark = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "dark"])
male_characters_hair_dull = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "dull"])
male_characters_hair_ginger = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "ginger"])
male_characters_hair_white = len(reading_xlxs_hp_male[reading_xlxs_hp_male["Hair_Colour"] == "white"])



# FEMALES
fem_characters_hair_bald = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "bald"])
fem_characters_hair_black = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "black"])
fem_characters_hair_blond = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "blond"])
fem_characters_hair_brown = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "brown"])
fem_characters_hair_dark = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "dark"])
fem_characters_hair_dull = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "dull"])
fem_characters_hair_ginger = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "ginger"])
fem_characters_hair_white = len(reading_xlxs_hp_female[reading_xlxs_hp_female["Hair_Colour"] == "white"])




df_plot_male = pd.DataFrame({
        'Bald' :  male_characters_hair_bald,
        'Black' : male_characters_hair_black,
        'Blonde' : male_characters_hair_blond,
        'Brown' : male_characters_hair_brown,
        'Dark' : male_characters_hair_dark,
        'Dull' : male_characters_hair_dull,
        'Ginger' : male_characters_hair_ginger,
        'White' : male_characters_hair_white
    
    }, index=[1])


df_plot_fem = pd.DataFrame({
        'Bald' :  fem_characters_hair_bald,
        'Black' : fem_characters_hair_black,
        'Blonde' : fem_characters_hair_blond,
        'Brown' : fem_characters_hair_brown,
        'Dark' : fem_characters_hair_dark,
        'Dull' : fem_characters_hair_dull,
        'Ginger' : fem_characters_hair_ginger,
        'White' : fem_characters_hair_white
    
    }, index=[1])




df_plot_male.plot(kind='bar')
df_plot_fem.plot(kind='bar')


```
